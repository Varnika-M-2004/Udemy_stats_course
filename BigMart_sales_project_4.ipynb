```python
import pandas as pd
import numpy as np
```

```python
sales_data = pd.read_csv("/content/bigmartsales.csv")
sales_data.head()
```

**Calculate mean, median, mode for Item_MRP column.**

```python
sales_data.columns
```

```python
print("Mean of Item_MRP column is:",sales_data['Item_MRP'].mean())
print("Median of Item_MRP column is:",sales_data['Item_MRP'].median())
print("Mode of Item_MRP column is:",sales_data['Item_MRP'].mode())
```

The above output concludes that:
- The average of the Item_MRP column is 140.99 ~ 141. Which means that the average MRP of items is about Rs.141/-
- The median of the Item_MRP column is 143 which means that approximately half of the items MRP is less than Rs.143/- and the other half is greater than Rs.143/-
- The mode of the column is 172 which means that the most frequently occurring item MRP is Rs.172/-
- As the mean and median values are almost equal, we can say that the distribution is almost symmetrical.

**Calculate the measure of spread - range, IQR, variance, std for Item_MRP column.**

```python
from scipy.stats import iqr
print("Range of Item_MRP column is:",np.ptp(sales_data['Item_MRP']))
print("IQR of Item_MRP column is:",iqr(sales_data['Item_MRP']))
print("Variance of Item_MRP column is:",sales_data['Item_MRP'].var())
print("Standard Deviation of Item_MRP column is:",sales_data['Item_MRP'].std())
```

The above output suggests that the Range of the Item_MRP column is vast which means that the item MRPs are varied. This is proved by calculating the variance of the column. The variance value is 3878 which indicates that the MRP of items si spread out vastly and has lot of variety.

**Obtain different summary statistics (skewness and kurtosis) for the Item_Outlet_Sales column.**

```python
from scipy.stats import skew, kurtosis
print("Skewness of the Item_Outlet_Sales column is:",skew(sales_data['Item_Outlet_Sales']))
print(f"Pearsons kurtosis of the Item_Outlet_Sales column is: {kurtosis(sales_data['Item_Outlet_Sales'], fisher = False)}")
print(f"Fisher's kurtosis of Item_Outlet_Sales column is: {kurtosis(sales_data['Item_Outlet_Sales'], fisher = True)}")
```

The above output suggests that:
- As the skewness is 1.177, it indicates that the Item_Outlet_Sales column is positively skewed (right-skewed).
- As the Pearsons kurtosis value is 4.614 which is greater than 3 indicates a leptokurtic distribution i.e., a distribution with heavier tails and a sharper peak.

**Detect and remove outliers based on the data in Item_MRP and Item_Outlet_Sales columns using z-score.**

```python
from scipy.stats import zscore
print("Z-score of Item_MRP and Item_Outlet_Sales column:")
z = abs(zscore(sales_data[['Item_MRP', 'Item_Outlet_Sales']], axis=0))
print(pd.DataFrame(z))
```

```python
outliers = np.where(z>3)
outliers
```

The above output indicates the row values which exceed the given threshold value of 3 and hence are considered as outliers.

```python
np.size(outliers[0])
```

This output shows the total count of the outliers which we have to remove from the main data.

```python
sales_data = sales_data.drop(outliers[0])
```

This piece of code is written to drop the outliers from the original data.

```python
sales_data
```

**Detect outliers based on data in Item_Outlet_Sales column using IQR.**

```python
Q1 = sales_data['Item_Outlet_Sales'].quantile(0.25)
Q3 = sales_data['Item_Outlet_Sales'].quantile(0.75)
IQR = Q3 - Q1
print("IQR of Item_Outlet_Sales column is:",IQR)
outlier_1 = sales_data['Item_Outlet_Sales'] < Q1 - 1.5*IQR
outlier_2 = sales_data['Item_Outlet_Sales'] > Q3 + 1.5*IQR
total_outliers = sales_data['Item_Outlet_Sales'][(outlier_1) | (outlier_2)]
print("Outliers based on IQR:")
print(total_outliers)
```

The above output depicts that there are 113 rows which are considered as outliers based on the given formula. Hence, we need to drop them.

```python
sales_data = sales_data.drop(total_outliers.index)
sales_data
```

The above output gives a total of 8320 rows as it has removed the 113 rows that are considered as outliers in the previous cells.
