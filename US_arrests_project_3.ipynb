```python
import pandas as pd
import numpy as np
```

```python
us_arrest_data = pd.read_excel("/content/USArrests.xlsx")
us_arrest_data.head()
```

**Analyze measures of spread and shape (kurtosis, skewness) on Assault and Fraud columns.**

```python
#Range
print("Range of Assault column:",np.ptp(us_arrest_data['Assault']))
print("Range of Fraud column:",np.ptp(us_arrest_data['Fraud']))

#IQR
Q1_assault = us_arrest_data['Assault'].quantile(0.25)
Q3_assault = us_arrest_data['Assault'].quantile(0.75)
IQR_assault = Q3_assault - Q1_assault
print("IQR of Assault column:",IQR_assault)

Q1_fraud = us_arrest_data['Fraud'].quantile(0.25)
Q3_fraud = us_arrest_data['Fraud'].quantile(0.75)
IQR_fraud = Q3_fraud - Q1_fraud
print("IQR of Fraud column:",IQR_fraud)

#Variance
print("Vaiance of Assault column:",us_arrest_data['Assault'].var())
print("Variance of Fraud column:",us_arrest_data['Fraud'].var())

#Standard Deviation
print("Standard Deviation of Assault column:",us_arrest_data['Assault'].std())
print("Standard Deviation of Fraud column:",us_arrest_data['Fraud'].std())
```

The above output indicates that teh range of Assault column is much greater than the Fraud column and hence the data in Assault column is greatly varied and distributed than Fraud column.

```python
from scipy.stats import skew, kurtosis
print("Skewness of Assault column:",skew(us_arrest_data['Assault']))
print("Skewness of Fraud column:",skew(us_arrest_data['Fraud']))

print(f"Pearson kurtosis of Assault column: {kurtosis(us_arrest_data['Assault'], fisher = False)}")
print(f"Fisher's kurtosis of Assault column: {kurtosis(us_arrest_data['Assault'], fisher = True)}")

print(f"Pearson kurtosis of Fraud column: {kurtosis(us_arrest_data['Fraud'], fisher = False)}")
print(f"Fisher's kurtosis of Fraud column: {kurtosis(us_arrest_data['Fraud'], fisher = True)}")
```

- Skewness: 0.2273178665911123
A skewness value of 0.227 indicates that the data is slightly positively skewed (right-skewed). This means there is a slight asymmetry with a longer tail on the right side of the distribution, but it is close to being symmetrical.
Fraud Column:

- Skewness: 0.7769613202124208
A skewness value of 0.777 indicates that the data is moderately positively skewed (right-skewed). This suggests a more noticeable asymmetry with a longer tail on the right side of the distribution.
Kurtosis
Assault Column:

- Pearson Kurtosis: 1.930979948426923
Fisher's Kurtosis: -1.069020051573077
Pearson kurtosis of 1.931 (which is less than 3) and Fisher's kurtosis of -1.069 (since Fisher's kurtosis is calculated as Pearson kurtosis minus 3) indicate that the distribution is platykurtic. This means the distribution has thinner tails and a flatter peak compared to a normal distribution. There are fewer extreme outliers than a normal distribution.
Fraud Column:

- Pearson Kurtosis: 3.201897793084265
Fisher's Kurtosis: 0.20189779308426514
Pearson kurtosis of 3.202 (which is slightly greater than 3) and Fisher's kurtosis of 0.202 indicate that the distribution is slightly leptokurtic. This means the distribution has slightly fatter tails and a sharper peak compared to a normal distribution. There are more extreme outliers than a normal distribution, but the difference is not very pronounced.

**Detect outliers based on data in UrbanPop column using IQR.**

```python
Q1_urbanpop = us_arrest_data['UrbanPop'].quantile(0.25)
Q3_urbanpop = us_arrest_data['UrbanPop'].quantile(0.75)
IQR_urbanpop = Q3_urbanpop - Q1_urbanpop
print("IQR of UrbanPop column is:",IQR_urbanpop)

outlier_1 = us_arrest_data['UrbanPop'] < Q1_urbanpop - 1.5*IQR_urbanpop
outlier_2 = us_arrest_data['UrbanPop'] > Q3_urbanpop + 1.5*IQR_urbanpop
print("Outliers in UrbanPop column are:",us_arrest_data['UrbanPop'][outlier_1 |outlier_2])
```

From the above output we can conclude that the UrbanPop column has only 1 outlier i.e., row 20 and the IQR of UrbanPop is 23.25.

**Detect outliers based on UrbanPop column using z-score.**

```python
from scipy.stats import zscore
print("UrbanPop column z-score:")
z = pd.DataFrame(zscore(us_arrest_data['UrbanPop']))
z
```

```python
print(np.where(z > 1))
```

In the above code cell, we have mentioned a specific threshold value for detecting outliers i.e., if the zscore value is gretaer than 1, it'll be considered as an outlier. Hence, row 20 has a zscore value greater than 1 and is an outlier.

**Calculate Range, variance, and Standard Deviation of data in UrbanPop and Assault column.**

```python
print("Range of UrbanPop column:",np.ptp(us_arrest_data['UrbanPop']))
print("Range of Assault column:",np.ptp(us_arrest_data['Assault']))
print("Variance of UrbanPop column:",us_arrest_data['UrbanPop'].var())
print("Variance of Assault column:",us_arrest_data['Assault'].var())
print("Standard Deviation of UrbanPop column:",us_arrest_data['UrbanPop'].std())
print("Standard Deviation of Assault column:",us_arrest_data['Assault'].std())
```
